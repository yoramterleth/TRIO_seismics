**As we saw earlier, the seismometer just records the motion of the ground over time.** So we're going to have to look at a plot of the ground displacement over time. 
The seismometer recorded at **100 Hz**, which means it made **100 measurements each second**, how many measurements will there be in an hour?


- Fill in the values, then click *Run* above to run this python code. 

```python
# number of measuremens per second:
measurement_rate = 0

# number of seconds per hour 
sec_hour = 0

# number of measurements in an hour 
nb_measurements = measurement_rate * sec_hour

# print the result
print('There are '+ str(nb_measurements)+ ' measurements per hour.')
```

Now we will load the data. We look at one hour of data, measured on September 6th 2020. To check what we calculated above, let's print the number of measurements in the data set. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the CSV file
# Replace 'seismic_data.csv' with your actual CSV file path
data = np.genfromtxt('seismic_data.csv', delimiter=',', names=True, dtype=None, encoding=None)

# Extract the time and data vectors
time = data['Time']
seismic_data = data['Data']

# Convert the time vector from string to datetime objects if necessary
# If the time vector is in a string format that can be directly plotted, you can skip this step
# This step assumes that the time is in ISO format
from datetime import datetime
time = np.array([datetime.strptime(t,'%Y-%m-%dT%H:%M:%S.%fZ') for t in time])

print('There is a total of '+ str(len(seismic_data))+' measurements in the data set.')
```

**Now, let's finally plot the data!**

```python
# Plot the data
plt.figure(figsize=(20, 12))
plt.plot(time, seismic_data, label='Seismic Data', color='black')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Seismic Data Over Time - Vertical channel - 50 Hz')
plt.legend()
plt.grid(True)
plt.show()
```
