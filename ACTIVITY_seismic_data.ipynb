Now that we have looked at the seismometer, it's time to look at some seismic data from a glacier! 

We're going to investgate data from a station that was installed on a glacier in Alaska, called S\'it' Kus\'a. This means Narrow Glacier in Tlingit. 

This is a map of the glacier:

![Map of the glacier](./TRIO/IMG_4260.png)

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the CSV file
# Replace 'seismic_data.csv' with your actual CSV file path
data = np.genfromtxt('seismic_data.csv', delimiter=',', names=True, dtype=None, encoding=None)

# Extract the time and data vectors
time = data['Time']
seismic_data = data['Data']

# Convert the time vector from string to datetime objects if necessary
# If the time vector is in a string format that can be directly plotted, you can skip this step
# This step assumes that the time is in ISO format
from datetime import datetime
time = np.array([datetime.fromisoformat(t) for t in time])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, seismic_data, label='Seismic Data', color='black')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Seismic Data Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
